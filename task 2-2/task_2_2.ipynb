```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("medical appointment dataset.csv")
print("Dataset loaded successfully!")

print("First 5 Rows")
display(df.head())

print("Last 5 Rows")
display(df.tail())

print("Dataset Shape (Before Cleaning):", df.shape)

missing_values = df.isnull().sum()
print("Missing Values in Each Column:")
display(missing_values)

plt.figure(figsize=(8,4))
plt.bar(missing_values.index, missing_values.values)
plt.xticks(rotation=90)
plt.title("Missing Data Pattern (Before Cleaning)")
plt.xlabel("Columns")
plt.ylabel("Missing Count")
plt.show()

if missing_values.sum() > 0:
    missing_values[missing_values > 0].plot(
        kind="bar",
        title="Missing Values Count per Column",
        figsize=(8,4)
    )
    plt.xlabel("Columns")
    plt.ylabel("Missing Count")
    plt.show()
else:
    print("No missing values found in the dataset")

numerical_cols = df.select_dtypes(include=["int64", "float64"]).columns
categorical_cols = df.select_dtypes(include=["object"]).columns

print("Numerical Columns:", list(numerical_cols))
print("Categorical Columns:", list(categorical_cols))

for col in numerical_cols:
    if df[col].isnull().sum() > 0:
        df[col].fillna(df[col].median(), inplace=True)

for col in categorical_cols:
    if df[col].isnull().sum() > 0:
        df[col].fillna(df[col].mode()[0], inplace=True)

threshold = len(df) * 0.5
df = df.dropna(thresh=threshold, axis=1)

print("Columns removed based on missing value threshold")

print("Missing Values After Cleaning:")
display(df.isnull().sum())

print("Dataset Shape (After Cleaning):", df.shape)

print("Dataset Info After Cleaning:")
df.info()

print("Statistical Summary After Cleaning:")
display(df.describe())

df.to_csv("cleaned_medical_appointment_dataset.csv", index=False)
print("Cleaned dataset saved successfully!")
```
